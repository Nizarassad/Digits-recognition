```python
import os
import pandas as pd
from PIL import Image

# Function to create folders for each digit
def create_digit_folders(output_directory):
    for i in range(10):
        digit_folder = os.path.join(output_directory, str(i))
        if not os.path.exists(digit_folder):
            os.makedirs(digit_folder)

# Function to convert CSV data to images and save them in respective folders
def csv_to_images(csv_file, output_directory):
    df = pd.read_csv(csv_file)

    # Get the column name containing labels
    label_column = df.columns[0]

    # Create folders for each digit
    create_digit_folders(output_directory)

    for idx, row in df.iterrows():
        label = row[label_column]
        pixel_values = row.drop(label_column).values
        image_data = pixel_values.reshape(28, 28).astype('uint8')

        image = Image.fromarray(image_data)

        # Save the image in the corresponding digit folder
        image_path = os.path.join(output_directory, str(label), f"{idx}.png")
        image.save(image_path)

if __name__ == "__main__":
    # Replace 'mnist_data.csv' with the path to your MNIST CSV file
    csv_file_path = '/content/mnist_test.csv'

    # Replace 'output_folder' with the path to the folder where you want to save the images
    output_folder_path = '/content/mnistdigittestdataset'

    csv_to_images(csv_file_path, output_folder_path)
```
